# 5.

# 6.

```python
dr_grouped = dr.groupby('Beverage_prep')[['Calories']].mean()

av_calories = drinks.describe()['Calories']['mean']

dr_grouped[dr_grouped['Calories'] > av_calories]
```

# 7.

```python
coffees = drinks[drinks['Beverage'] == 'Brewed Coffee']

min_cal = coffees.describe()['Calories']['min']

coffees[coffees['Calories'] == min_cal]
```
